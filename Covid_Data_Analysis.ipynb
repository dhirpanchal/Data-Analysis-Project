 # Finding which continent has maximum human development index'

```python
df[["continent",'human_development_index']].value_counts(ascending=False).head(1)
```

# Finding which continent has minimum gdp per capita

```python
df[['continent','gdp_per_capita']].value_counts(ascending=True).head(1)
```

# Filtering the dataframe with only these columns

```python
df=df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df
```

# Removing duplicate values

```python
df.drop_duplicates()
```

# Finding Null Values

```python
df.isnull().sum()
```

# Removing all observations where continent column value is missing

```python
df1=df.dropna(subset=['continent'])
df1
```

```python
df1.isnull().sum()
```

#  Filling all missing values with 0

```python
df.fillna(0)
```

# Converting date column in datetime format

```python
df.dtypes
df.info()
```

```python
df['date']= pd.to_datetime(df['date'])
```

```python
df.dtypes
```

# New column month after extracting month data from date

```python
df['Month'] = pd.DatetimeIndex(df['date']).month
df.head()
```

# Finding max value in all columns 

```python
df_groupby=df.groupby(['continent']).max().reset_index()
df_groupby
```

# Creating a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df['total_deaths_to_total_cases']=df["total_deaths"]/df['total_cases']
df
```
